```python
import pandas as pd
import os
```

```python

files = [file for file in os.listdir("./Sales_Data")]
all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("./Sales_Data/"+ file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data.to_csv("all_data.csv",index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python

all_data["Month"]= all_data["Order Date"].str[0:2]
all_data = all_data.dropna()
##to display the 'Or' value in data 
all_data = all_data[all_data["Order Date"].str[0:2] != 'Or']
all_data["Month"]= all_data["Month"].astype(int)
all_data.head()
```

```python
## to correct the type of other variables
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
## 1to find out the best month of sales
all_data.groupby('Month').sum()
#to store the value in highest in result 
result = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
month = range(1,13)
plt.bar(month,result['Sales'])
plt.xticks(month)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

```python
#2which city has highest sales
#apply() is used to run any function in pandas
def get_city(address):
    return address.split(',')[1]
def get_pin(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x :get_city(x) + ' ' + get_pin(x))
all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt
cities = [city for city , df in all_data.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation ='vertical', size=8)
plt.ylabel('sales in usd')
plt.xlabel('city name')
plt.show()
```

```python
#3 what is the time to display the aadvertisment to inc sales
#using datetime lib to find out
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()
```

```python
hours = [hour for hour , df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.ylabel('no. of orders')
plt.xlabel('hour')
plt.grid()
plt.show()
```

```python
#4 which product sold most often
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
count = Counter()
for raw in df['Grouped']:
    raw_list = raw.split(',')
    count.update(Counter(combinations(raw_list,2)))
for key ,value in count.most_common(10):
    print(key,value)
```
